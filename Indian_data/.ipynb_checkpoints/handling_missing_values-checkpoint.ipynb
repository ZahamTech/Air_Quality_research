## Data Handling

```python
df.isnull().sum()
```

```python
df.query('aqi != aqi')
```

```python
import plotly.express as px

fig = px.line(df, x='timestamp_local', y='aqi', title='AQI with Slider')

fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
df['timestamp_local'] = pd.to_datetime(df['timestamp_local'])
```

```python
df = df.set_index('timestamp_local')
```

```python
df.loc['2022-06-30':'2022-07-03']
```

```python
df_na = df.copy()
df_na= df_na.dropna()
```

```python
pd.plotting.autocorrelation_plot(df_na['2022':'2024']['aqi'])
```

```python
df_imp = df.copy()
```

```python
df_imp.head()
```

```python
df_imp.drop(columns=['temp','city_name','country_code','co','no2','o3','pm10','pm25','so2'],inplace=True)
```

```python
df_imp
```

```python
df_imp['aqi_FFILL']=df_imp['aqi'].ffill()
```

```python
df_imp['aqi_BFILL']=df_imp['aqi'].bfill()
```

```python
df_imp['aqi'].rolling(window=2, min_periods=1).mean()
```

```python
df_imp['aqi_ROLLING']=df_imp['aqi'].rolling(window=80, min_periods=1).mean()
```

```python
df_imp['aqi_interpolate_p'] = df_imp['aqi'].interpolate(method='spline', axis=0, order=2)
```

```python
df_imp['aqi_interpolate_s'] = df_imp['aqi'].interpolate(method='nearest', axis=0)
```

```python
df_imp
```
