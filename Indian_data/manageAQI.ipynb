```python
import pandas as pd
import numpy as np
```

```python
# Function to calculate AQI based on concentration
def calculate_aqi(cp, breakpoints):
    for (blo, bhi, ilo, ihi) in breakpoints:
        if blo <= cp <= bhi:
            return ((ihi - ilo) / (bhi - blo)) * (cp - blo) + ilo
    return np.nan  # Return NaN if the value is out of range
```

```python
pm25_breakpoints = [
    (0, 30, 0, 50),        # Good
    (31, 60, 51, 100),     # Satisfactory
    (61, 90, 101, 200),    # Moderate
    (91, 120, 201, 300),   # Poor
    (121, 250, 301, 400),  # Very Poor
    (251, float('inf'), 401, 500)  # Severe
]

co_breakpoints = [
    (0, 1, 0, 50),         # Good
    (1.1, 2, 51, 100),     # Satisfactory
    (2.1, 10, 101, 200),   # Moderate
    (10.1, 17, 201, 300),  # Poor
    (17.1, 34, 301, 400),  # Very Poor
    (34.1, float('inf'), 401, 500)  # Severe
]

nox_breakpoints = [
    (0, 40, 0, 50),        # Good
    (41, 80, 51, 100),     # Satisfactory
    (81, 180, 101, 200),   # Moderate
    (181, 280, 201, 300),  # Poor
    (281, 400, 301, 400),  # Very Poor
    (401, float('inf'), 401, 500)  # Severe
]
no_breakpoints = [
    (0, 40, 0, 50),        # Good
    (41, 80, 51, 100),     # Satisfactory
    (81, 180, 101, 200),   # Moderate
    (181, 280, 201, 300),  # Poor
    (281, 400, 301, 400),  # Very Poor
    (401, float('inf'), 401, 500)  # Severe
]

# Updated breakpoints for NO2
no2_breakpoints = [
    (0, 40, 0, 50),        # Good
    (41, 80, 51, 100),     # Satisfactory
    (81, 180, 101, 200),   # Moderate
    (181, 280, 201, 300),  # Poor
    (281, 400, 301, 400),  # Very Poor
    (401, float('inf'), 401, 500)  # Severe
]

# Updated breakpoints for O3
o3_breakpoints = [
    (0, 50, 0, 50),        # Good
    (51, 100, 51, 100),    # Satisfactory
    (101, 168, 101, 200),  # Moderate
    (169, 208, 201, 300),  # Poor
    (209, 748, 301, 400),  # Very Poor
    (749, float('inf'), 401, 500)  # Severe
]
# Updated breakpoints for SO2
so2_breakpoints = [
    (0, 40, 0, 50),        # Good
    (41, 80, 51, 100),     # Satisfactory
    (81, 380, 101, 200),   # Moderate
    (381, 800, 201, 300),  # Poor
    (801, 1600, 301, 400), # Very Poor
    (1601, float('inf'), 401, 500)  # Severe
]
```

```python
## AQI bucketing
def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN
```

```python
# Apply AQI calculations for each pollutant in the dataset
df['AQI_PM25'] = df['pm25'].apply(lambda x: calculate_aqi(x, pm25_breakpoints))
df['AQI_NOX'] = df['nox'].apply(lambda x: calculate_aqi(x, nox_breakpoints))
df['AQI_NO'] = df['no'].apply(lambda x: calculate_aqi(x, no_breakpoints))
df['AQI_CO'] = df['co'].apply(lambda x: calculate_aqi(x, co_breakpoints))
df['AQI_NO2'] = df['no2'].apply(lambda x: calculate_aqi(x, no2_breakpoints))
df['AQI_O3'] = df['o3'].apply(lambda x: calculate_aqi(x, o3_breakpoints))
df['AQI_SO2'] = df['so2'].apply(lambda x: calculate_aqi(x, so2_breakpoints))
```

```python
# Combine the highest AQI for overall AQI per row
df['AQI'] = df[['AQI_PM25', 'AQI_NOX','AQI_NO','AQI_CO', 'AQI_NO2', 'AQI_O3', 'AQI_SO2']].max(axis=1)
df["AQI_Category"] = df["AQI"].apply(lambda x: get_AQI_bucket(x))
# Save the result to a new CSV file
df.to_csv("df_AQI.csv", index=False)
```
